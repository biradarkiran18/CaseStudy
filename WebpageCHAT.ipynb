```python
import requests
from bs4 import BeautifulSoup

def extract_relevant_text(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')

        content = [p.get_text() for p in soup.find_all("p")]

        return "\n".join(content)
    except requests.exceptions.RequestException as e:
        return f"Error fetching the webpage: {e}"
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

def find_answer(question, context):
    if "Error" in context:
        return context

    result = qa_pipeline(question=question, context=context)

    if result['score'] > 0.3:
        return result['answer']
    else:
        return "I don't know the answer."
```

```python
# Define the URL to fetch content from
url = "https://en.wikipedia.org/wiki/Generative_artificial_intelligence"

question = input("Enter the question to ask: ")

context = extract_relevant_text(url)

answer = find_answer(question, context)

print("Question:", question)
print("Answer:", answer)
```
